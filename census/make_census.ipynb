```python
import pandas as pd
import nameutils as nu
```

```python
# 2016 census year was late.
census_2016_csv = 'src/2016 Census_2013 Federal districts.csv'
src_df = pd.read_csv(census_2016_csv)
src_df.dropna(inplace=True)
for col in src_df.columns:
    if col not in ['FED_Name', 'Average number of persons per census family']:
        src_df[col] = src_df[col].astype('int64')
nu.rename_cols(src_df)
```

```python
src_df.info()
src_df.to_csv('2016_2013_Census.csv', index=False)
#for c in src_df.columns: print(c)
```

```python
input_lst = [
    '2001_1996_Census.csv', '2001_2003_Census.csv', '2006_2003_Census.csv', '2011_2003_Census.csv', '2011_2013_Census.csv', '2016_2013_Census.csv'
]

df_lst = []
for csv in input_lst:
    df = pd.read_csv(csv)
    name = csv.split('.')[0]
    df.name = name
    #nu.rename_cols(df)
    df_lst.append(df)

res = pd.concat(df_lst, join='inner', ignore_index=True, sort=False)
res.info()
#for c in res.columns: print(c)
```

```python
table_name = "Census"
query = nu.gen_create_table(res, table_name)
print(query)
with open('sql/' + table_name + ".sql", "w") as sql_file:
    sql_file.write(query)
res.to_csv(table_name + '.csv', index=False)
```

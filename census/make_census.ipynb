```python
import pandas as pd
import nameutils as nu
```

```python
input_lst = [
    '2001_1996_Census.csv', '2001_2003_Census.csv', '2006_2003_Census.csv', '2011_2003_Census.csv', '2011_2013_Census.csv'
]

df_lst = []
for csv in input_lst:
    df = pd.read_csv(csv)
    name = csv.split('.')[0]
    df.name = name
    #nu.rename_cols(df)
    df_lst.append(df)

res = pd.concat(df_lst, join='inner', ignore_index=True, sort=False)
res.info()
```

```python
table_name = "Census"
query = nu.gen_create_table(res, table_name)
print(query)
with open('sql/' + table_name + ".sql", "w") as sql_file:
    sql_file.write(query)
res.to_csv(table_name + '.csv', index=False)
```

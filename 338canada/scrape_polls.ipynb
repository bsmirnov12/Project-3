```python
import pandas as pd
```

```python
url = 'https://338canada.com/districts.htm'
```

```python
tables = pd.read_html(url)
feds_s = tables[0]['Electoral district'].map(lambda s: s.split(' ')[0])
proj_s = tables[0]['Latest projection'].map(lambda s: s.split(' ')[0])
proj_df = pd.DataFrame([feds_s, proj_s]).transpose()
```

```python
toss_df = proj_df.loc[proj_df['Latest projection'] == 'Toss']
```

```python
toss_df
```

```python
results_lst = ['LPC', 'BQ',  'LPC', 'LPC', 'BQ',  'LPC', 'LPC', 'BQ',  'CPC', 'LPC', 'LPC', 'CPC', 'LPC', 'CPC', 'CPC',
               'LPC', 'LPC', 'CPC', 'LPC', 'LPC', 'LPC', 'NDP', 'CPC', 'LPC', 'CPC', 'CPC', 'CPC', 'LPC', 'NDP', 'CPC',
               'LPC']
```

```python
proj_df.loc[proj_df['Latest projection'] == 'Toss', 'Latest projection'] = results_lst
```

```python
proj_df.rename(columns={'Electoral district': 'FED Id', 'Latest projection': 'Winning Pid'}, inplace=True)

rename_dct = {
    'LPC': 'LIB',
    'CPC': 'CPC',
    'NDP': 'NDP',
    'BQ': 'BQ',
    'GPC': 'GRN',
    'PPC': 'OTH',
    'IND': 'OTH'
}

proj_df['Winning Pid'] = proj_df['Winning Pid'].map(lambda x: rename_dct[x])
```

```python
proj_df.to_csv('projection.csv', index=False)
```

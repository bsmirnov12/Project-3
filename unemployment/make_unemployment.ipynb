```python
# Depedences
import pandas as pd
```

```python
# Constants
input_csv = 'unemployment_2000-2019_raw.csv'
output_csv = 'unemployment_2000-2019.csv'
```

```python
# Dataset with per candidate per riding results
df = pd.read_csv(input_csv)
```

```python
# Rename columns
rename_col = {
    'Geography2': 'Geography',
    'Reference period': 'Year'
}
df.rename(columns=rename_col, inplace=True)

# Leave only election years
df.drop(df.loc[~df['Year'].isin([2000, 2004, 2006, 2008, 2011, 2015, 2019])].index, inplace=True)

# Fill in Geography gaps
df['Geography'] = df['Geography'].interpolate(method='pad')
df.reset_index(inplace=True, drop=True)

# Save Canada wide stats
canada_df = df[df['Geography'] == 'Canada'].copy()

# Append missing data using Canada wide values as default
for prov in ['Yukon', 'Northwest Territories', 'Nunavut']:
    canada_df['Geography'] = prov
    df = df.append(canada_df, ignore_index=True)

# Remove Canada wide stats
df.drop(index=range(0, 7), inplace=True)

# Replace province name with province Id and rename the column accordingly
prov_name_id = {
    'Alberta': 48,
    'British Columbia': 59,
    'Nunavut': 62,
    'Manitoba': 46,
    'New Brunswick': 13,
    'Newfoundland and Labrador': 10,
    'Nova Scotia': 12,
    'Ontario': 35,
    'Prince Edward Island': 11,
    'Quebec': 24,
    'Saskatchewan': 47,
    'Yukon': 60,
    'Northwest Territories': 61
}

df['Geography'] = df['Geography'].map(prov_name_id)
df.rename(columns={'Geography': 'Prov Id'}, inplace=True)

# Save the result
df.to_csv(output_csv, index=False)

df
```

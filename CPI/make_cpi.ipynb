```python
# Depedences
import pandas as pd
```

```python
# Constants
input_csv = 'cpi_2000-2019_raw.csv'
output_csv = 'cpi_2000-2019.csv'
```

```python
# Dataset with per candidate per riding results
cpi_df = pd.read_csv(input_csv)
```

```python
# Leave only election years
cpi_df.drop(cpi_df.loc[~cpi_df['Reference period'].isin([2000, 2004, 2006, 2008, 2011, 2015, 2019])].index, inplace=True)

# Fill in Geography gaps
cpi_df['Geography'] = cpi_df['Geography'].interpolate(method='pad')

# Set new index by geography and year
cpi_df.set_index(['Geography', 'Reference period'], inplace=True)
```

## Filling in the gaps

```python
# This is lame, but I don't have time to find out how make the assignment in one command.
# This doesn't work:
# cpi_df.loc[('Iqaluit, Nunavut 8', slice(None))] = cpi_df.loc[('Canada', slice(None))]

missing_data = [
    'Rented accommodation',
    "Homeowners' replacement cost",
    'Property taxes and other special charges',
    "Homeowners' home and mortgage insurance 5",
    "Homeowners' maintenance and repairs",
    'Natural gas'
]

for year in [2000, 2004, 2006, 2008, 2011, 2015, 2019]:
    # Nunavut data is missing entirely
    cpi_df.loc[('Iqaluit, Nunavut 8', year)] = cpi_df.loc[('Canada', year)]

    # Alberta missing 'Fuel oil and other fuels' indexes
    cpi_df.loc[('Alberta', year), 'Fuel oil and other fuels'] = cpi_df.loc[('Canada', year), 'Fuel oil and other fuels']
    
    # No 'Natural gas' info for the Atlantic Provinces
    cpi_df.loc[(['New Brunswick', 'Newfoundland and Labrador', 'Nova Scotia', 'Prince Edward Island'], year), 'Natural gas'] = cpi_df.loc[('Canada', year), 'Natural gas']

    # No housing info and 'Natural gas' for all northern territories
    #This doesn't work: cpi_df.loc[(['Whitehorse, Yukon 7', 'Yellowknife, Northwest Territories 7'], year), missing_data] = cpi_df.loc[('Canada', year), missing_data]
    for col in missing_data:
        cpi_df.loc[(['Whitehorse, Yukon 7', 'Yellowknife, Northwest Territories 7'], year), col] = cpi_df.loc[('Canada', year), col]

```

```python
# 'Internet access services' - missing data for 2000-2002
for geo in cpi_df.index.levels[0]:
    cpi_df.loc[(geo, 2000), 'Internet access services (200212=100) 6'] = cpi_df.loc[(geo, 2004), 'Internet access services (200212=100) 6']

#cpi_df.loc[(slice(None), [2000, 2004, 2006]), 'Internet access services (200212=100) 6']
```

## Finalizing dataframe for output

```python
# Dropping country level
cpi_df.drop('Canada', level=0, inplace=True)
```

```python
# Replace province name with province Id
prov_name_id = {
    'Alberta': 48,
    'British Columbia': 59,
    'Iqaluit, Nunavut 8': 62,
    'Manitoba': 46,
    'New Brunswick': 13,
    'Newfoundland and Labrador': 10,
    'Nova Scotia': 12,
    'Ontario': 35,
    'Prince Edward Island': 11,
    'Quebec': 24,
    'Saskatchewan': 47,
    'Whitehorse, Yukon 7': 60,
    'Yellowknife, Northwest Territories 7': 61
}

cpi_df.reset_index(inplace=True)
cpi_df['Geography'] = cpi_df['Geography'].map(prov_name_id)
```

```python
# Rename columns

rename_dct = {
    'Geography': 'Prov Id',
    'Reference period': 'Year',
    'Fresh or frozen meat (excluding poultry)': 'Fresh or frozen meat',
    'Cereal products (excluding baby food)': 'Cereal products',
    'Food purchased from restaurants 4': 'Food purchased from restaurants',
    "Homeowners' replacement cost": 'Homeowners replacement cost',
    "Homeowners' home and mortgage insurance 5": 'Homeowners home and mortgage insurance',
    "Homeowners' maintenance and repairs": 'Homeowners maintenance and repairs',
    'Internet access services (200212=100) 6': 'Internet access services',
    "Tobacco products and smokers' supplies": 'Tobacco products and smokers supplies'
}

cpi_df.rename(columns=rename_dct, inplace=True)
```

```python
# Saving cleaned dataset
cpi_df.to_csv(output_csv, index=False)
```
